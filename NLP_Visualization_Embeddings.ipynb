IMPLEMENTING DIMENSIONALITY REDUCTION
- REDUCING NUMBER OF FEATURES WHILE PRESERVING IMPORTANT INFO
1. WE HAVE TO REDUCE DIMENSIONS FROM 100D TO 2D -> as we cannot visualize 100D
THIS LET US VISUALLY SEE
1. SIMILAR WORDS AND CLUSTER TOGETHER
2. OPPOSITE WORDS SEPARATE
3. SEMANTIC RELATIONSHIPS

we can reduce dimensions by using:
1. PCA-linear reduction
-PCA finds the direction where data varies the most
-Directions=Principal Components
Suppose students:
-height and weight are correlated
-Instead of 2 features -PCA create 1 new feature representing both
Steps of PCA:
1. Standardize data
2. Compute covariace matrix
3. find eigen vector
4. select top components
5. Project data
2. t-SNE-Non Linear
3. UMAP

```python
!pip install gensim
import gensim.downloader as api
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
model=api.load('glove-wiki-gigaword-100')
print("Model loaded successfully")
```

```python
words = [
    "king","queen","man","woman","dog","cat","animal","car","truck","vehicle",
    "happy","sad","joy","anger"
]
vectors = [model[word] for word in words]
```

```python
pca = PCA(n_components=2)
reduced= pca.fit_transform(vectors)
```

```python
plt.figure(figsize=(10,6))
for i, word in enumerate(words):
  x=reduced[i][0]
  y=reduced[i][1]
  plt.scatter(x,y)
  plt.text(x+0.01,y+0.01,word)
  plt.title('word embedding PCA')
plt.show()
```

```python
from sklearn.manifold import TSNE
import numpy as np
vectors = np.array([model[word] for word in words])
tsne = TSNE(n_components=2,perplexity=3)
reduced_vectors = tsne.fit_transform(vectors)
plt.figure(figsize=(10,10))
for i, word in enumerate(words):
  x = reduced_vectors[i][0]
  y = reduced_vectors[i][1]
  plt.scatter(x,y)
  plt.text(x + 0.01, y+ 0.01, word)
plt.title("Word Embeddings TSNE")
plt.show()
```
